# Laser pulse hitting the probe in the middle
### This is the example shown in the README

Let us now consider a probe, which is in equilibrium, at temperature 0 in the beginning.
After some time a laser pulse (Gaussian shape in time and exponentially decaying in space $S(x,t)=g(t)*exp(-x)$) will hit the probe in the middle.



```python
from solpde import solpde
import numpy as np
import numericalunits as u
u.reset_units('SI')

h = solpde.heatpde()
```

Checking the initial default properties first

```python
h.getProperties()
```

In order to obtain the demanded dynamics, we will first change the default properties to the case in consideration.
* The boundary conditions on both sides are now Neumann conditions, i.e. $\frac{\partial_t\varphi(x_0,t)}{k(\varphi(x_0,t))}=\frac{\partial_t\varphi(x_N,t)}{k(\varphi(x_N,t))}=0$

* The Conductivity $k(\varphi)=1$ can be assumed to be constant
* Also the heat capacity $C(\varphi)=1$ will be assumed to be constant
* $\rho = 1$ since we are interested in general dynamics of the system. 
* S(x,t) =  $A\cdot e^{-(x-.5)^2\cdot \mu}\cdot e^{-(t-t_0)^2/(2s)}$

```python
h.Left_BC_Type = 1 #Neumann BC on both sides
h.Right_BC_Type = 1
h.Left_BC = lambda t: 0 # zero flux
h.Right_BC = lambda t: 0
h.heatCapacity = lambda phi: 1+0*phi#(455.2+0.112*phi)*(u.J/(u.kg*u.K))
h.conductivity = lambda phi:.1+0*phi#5.72*u.W/(u.m*u.K) #C(phi)
```

```python
s        = .02 
A        = 0.5
mu       = 200 
t0       = .3
h.source = lambda x,t: A*np.exp(-(x-.5)**2*mu)*np.exp(-(t-t0)**2/(2*s))
```

We can see a Gaussian shaped pulse in time, peaking at $t_0$, which is exponenitally decaying in space.

```python
h.surf_source()
```

Looking at a 2D countour and a 3D surface plot, we can depict the dynamics of the system

```python
h.spaceVStime()
h.surface()
```
