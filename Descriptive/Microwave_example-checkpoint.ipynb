# Microwave configuration
In this configuration we will consider a microwave like setting. Therefor we will set Dirichlet boundary conditions on both, the left and the right hand side of our probe. That is $\varphi(x_0,t)=\varphi(x_N,t)=0$. 
So the heat gets drained away at the edges, where it is fixet to a constant value. Then we will hit the probe with a uniforme source and see how the heat is distributed along the material. 


```python
from solvepde import solvepde
h = solvepde.heatpde()
```

Checking for the initial default conditions

```python
h.getProperties()
```

We can leave the firs few default properties as they are here now.
* This is, we leave the number of points in the xgrid and the length of our x space [0,1]
* we leave the time configruations as they are t [0,3] in steps of 0.0005
* also the density $\rho=1$ will stay the same in this case now

Now we will set specific simulation conditions for the equation under consideration:

$C(\varphi(x,t))\cdot\rho\frac{\partial\varphi(x,t)}{\partial t} = \frac{\partial}{\partial x}\left(k(\varphi(x,t))\cdot\frac{\partial\varphi(x,t)}{\partial x}\right) +S(x,t)$

where $k(\varphi)$ is the heat conductivity, $C(\varphi)$ is the heat capacity, $S(x,t)$ is the source hitting the probe, $\rho$ is the density and $\varphi$ is the temperature.

* `Left_BC_Type=0`& `Right_BC_Type=0` means that we are considering Dirichlet Boundary conditions
    in this case: $\varphi(x_0,t)=\varphi(x_N,t)=0$.
    This means, that the temperature on the left and right edge is staying constant.

* We set the heat capacity to a constant value $C(\varphi)=1$
* Also the conductivity $k(\varphi)=1$ in this case
* initially we consider the probe to be at an equilibrium: `h.init = 0`
    however this can in general also be a function of space $init(x)$: e.g.: `h.init = lambda x: 3 + 5*x`
    if one does not want to consider a uniforme initial heat distribution.
* The source is acting on the probe with a constant value `h.source = 1`. 
    Also this can be a function $S(x,t)$

```python
h.Left_BC_Type  = 0             #=>Dirichlet Boundary conditions on LHS
h.Right_BC_Type = 0             #=>Dirichlet Boundary conditions on RHS
h.Left_BC       = lambda t: 0   #setting the LHS edge to 0
h.Right_BC      = lambda t: 0   #setting the RHS edge to 0
h.heatCapacity  = lambda phi: 1 #constant capacity
h.conductivity  = lambda phi:.1 #constant conductivity
h.init          =  0            #initial temperature is 0
h.source        = 1             #setting a constant uniform source

h.getProperties()               #checking for the updated properties of object
```

Now, let us see the source as the majore influncial factore of the equation.
Therefor we use `h.surf_source()`
As expected, we see a uniformly distributed (in space and time) source acting on the entire grid.

```python
h.surf_source()
```

In order to see a 2D countour plot of the dynamics, one has to exicute the command `h.spaceVStime()`
Or simply `h.surface()` if one is interested in a 3D plot.

```python
h.spaceVStime()
h.surface()
```

Also it is possible to see the dynamics over time as an animation, when executing `h.animated(speed)`
`speed` is one required input argument for this call and determines the speed of the animation.

```python
speed = 30
h.animated(speed)
```
